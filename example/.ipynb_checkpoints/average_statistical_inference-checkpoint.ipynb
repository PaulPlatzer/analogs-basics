# Example

We will try to estimate a statistical relationship between variables X and Y using analogs. In particular, we will try to infer a new value of Y from the corresponding new observation of X, and a 'catalog' of joint, past observations of X and Y.

We will assume here that X is a random vector of dimension 3 and Y is of dimension 1. We assume a simple linear relationship between X and Y :

$$ Y = X_1 + 2 X_2 + \varepsilon ,$$

where $X$ and $\varepsilon$ are Gaussian random variables:

$$ X \sim \mathcal{N}( \;
        \left(
            \begin{matrix}
              0  \\
              0  \\
              0  \\
            \end{matrix}
        \right) \, , \,
        \left(
            \begin{matrix}
              1 \, 0 \, 0  \\
              0 \, 1 \, 0  \\
              0 \, 0 \, 1  \\
            \end{matrix}
        \right)\;  ) \, ,$$

$$ \varepsilon \sim \mathcal{N}( 1, 1 ) \, . $$

We then assume that we have a large catalog of joint observations of X and Y:

$$ \mathcal{C} = \left\lbrace (x_1,y_1) , \ldots , (x_N,y_N) \right\rbrace $$

and some new observations (100) of $X$ called "$x_\text{new}$" from which we want to infer the associated values "$y_\text{new}$".

$$ x_\text{new} = \text{known}, \;  y_\text{new} = \text{unknown} \,. $$

In other words, we are trying to estimate the conditional probability $P(Y|X)$.

### Drawing random data.

```python
N = 6000
x_cat = np.random.randn(N,3)
eps_cat = 1+1*np.random.randn(N)
y_cat = x_cat[:,0] + 2*x_cat[:,1] + eps_cat
```

```python
Nnew = 300
x_new = np.random.randn(Nnew,3)
eps_new = 1+1*np.random.randn(Nnew)
y_new = x_new[:,0] + 2*x_new[:,1] + eps_new
```

### Making the estimation using analogs.

Using analogs, the method to estimate $y_\text{new}$ is straightforward. For instance, trying to estimate the average value of the distribution P(Y|X) is done by taking the nearest neighbours (i.e., the analogs) of $x_\text{new}$ in the catalog, and averaging over the corresponding values of $y_\text{cat}$.

We call this estimation $\hat{y}_\text{new}$.

First, we must create a tree offline, which will help fasten the search for nearest-neigbours. This is particularly helpful if $N$ (the catalog's size) is large and if the dimension of the data where we make the search for neighbours (here 3, the dimension of X) is not too big (below 10).

See the documentation here https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KDTree.html

```python
kdt = KDTree(x_cat)
```

Now we search for the 5 nearest neighbours of each of the few instances of $x_\text{new}$ inside the catalog $\mathcal{C}$. In machine-learning terminology, $x_\text{new}$ is called a "query".

```python
dist, ind = kdt.query(x_new, k=5)
```

You can see that analogs are ordered by growing distance to the query.

```python
dist[:3] # displaying the 5 analogs of the 3 first values of x_new
```

Now let's computs $\hat{y}_\text{new}$ !

```python
y_hat = np.mean(y_cat[ind],axis=1)
```

```python
plt.figure()
plt.plot([np.min(y_new),np.max(y_new)],
        [np.min(y_new),np.max(y_new)], '.5')
plt.scatter(y_new, y_hat, alpha=.5)
plt.grid()
plt.title('5 analogs simple average')
plt.xlabel(r'$y_\mathrm{new}$')
plt.ylabel(r'$\hat{y}_\mathrm{new}$')
plt.show()
```

### Let's try again with more analogs.

```python
dist, ind = kdt.query(x_new, k=200)
y_hat = np.mean(y_cat[ind],axis=1)
```

```python
plt.figure()
plt.plot([np.min(y_new),np.max(y_new)],
        [np.min(y_new),np.max(y_new)], '.5')
plt.scatter(y_new, y_hat, alpha=.5)
plt.grid()
plt.title('200 analogs simple average')
plt.xlabel(r'$y_\mathrm{new}$')
plt.ylabel(r'$\hat{y}_\mathrm{new}$')
plt.show()
```

Raising the number of analogs slightly reduces the variability of $\hat{y}_\text{new}$, but it introduces spurious bias towards high and low values of $Y$. This is a typical case of __bias-variance trade-off__.


### Using kernels and weights

Let us try to remedy this problem using kernels, to give different weights to the analogs depending on their associated distances.

```python
dist, ind = kdt.query(x_new, k=200)
w = weights_from_kernel(dist)
y_hat = np.sum(y_cat[ind]*w,axis=1)
```

```python
plt.figure()
plt.plot([np.min(y_new),np.max(y_new)],
        [np.min(y_new),np.max(y_new)], '.5')
plt.scatter(y_new, y_hat, alpha=.5)
plt.grid()
plt.title('200 analogs and a Gaussian kernel')
plt.xlabel(r'$y_\mathrm{new}$')
plt.ylabel(r'$\hat{y}_\mathrm{new}$')
plt.show()
```

You can have a look at the weights, assessing the gap between the weights given to the closest and furthest analogs.

```python
w[0]
```

The default kernel is the Gaussian, it is pretty flat and does not make a big difference between "far" and "close" analogs. It seems like it is too flat to remedy the problem. Let us try another sharper kernel: the inverse kernel with $p=20$.

```python
dist, ind = kdt.query(x_new, k=200)
w = weights_from_kernel(dist, kernel = partial(inverse_kernel,p=20))
y_hat = np.sum(y_cat[ind]*w,axis=1)
```

```python
w[0]
```

```python
plt.figure()
plt.plot([np.min(y_new),np.max(y_new)],
        [np.min(y_new),np.max(y_new)], '.5')
plt.scatter(y_new, y_hat, alpha=.5)
plt.grid()
plt.title(r'200 analogs and an inverse kernel with $p=20$')
plt.xlabel(r'$y_\mathrm{new}$')
plt.ylabel(r'$\hat{y}_\mathrm{new}$')
plt.show()
```

There seems to be some improvement !

Note that the bandwidth could also be chosen by setting "lbda" in "weights_from_kernel".

```python
dist, ind = kdt.query(x_new, k=200)
w = weights_from_kernel(dist, kernel = partial(inverse_kernel,p=20),
                       lbda=.5)
y_hat = np.sum(y_cat[ind]*w,axis=1)
```

```python
w[:3]
```

```python
plt.figure()
plt.plot([np.min(y_new),np.max(y_new)],
        [np.min(y_new),np.max(y_new)], '.5')
plt.scatter(y_new, y_hat, alpha=.5)
plt.grid()
plt.title(r'200 analogs and an inverse kernel with $p=20$')
plt.xlabel(r'$y_\mathrm{new}$')
plt.ylabel(r'$\hat{y}_\mathrm{new}$')
plt.show()
```
